```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (10, 10)
```

```python
img = cv2.imread('scan.jpeg')
plt.imshow(img)
plt.axis('off');
```

# Getting table borders 

```python
#to get horizontal lines use horizontal structure element
horizontal = cv2.morphologyEx(img, cv2.MORPH_CLOSE, np.ones((1, 9))) 
#to get vertical lines use vertical structure element
vertical = cv2.morphologyEx(img, cv2.MORPH_CLOSE, np.ones((11, 1)))
# the resulting image is the minimum of both images
table = np.minimum(vertical, horizontal)
plt.imshow(table)
plt.axis('off');
```

# Deleting table borders 

```python
gray_table = cv2.cvtColor(table, cv2.COLOR_BGR2GRAY)
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
# deleting the table
gray_img_inv = cv2.bitwise_not(gray_img)
gray_table_inv = cv2.bitwise_not(gray_table)
inv_img_no_table = gray_img_inv - gray_table_inv
img_no_table = cv2.bitwise_not(inv_img_no_table)
plt.imshow(img_no_table, cmap = 'gray')
plt.axis('off');
```
